**Goal:** Make the model usable in real-world applications

---
## 📌 Step 1 — Build ML Pipeline (Preprocessing + Model)

```python
import pickle
import pandas as pd
import numpy as np

# Load saved pipeline from Sprint 3
with open('../models/final_model_pipeline.pkl', 'rb') as f:
    pipeline = pickle.load(f)

with open('../models/label_encoder.pkl', 'rb') as f:
    le_y = pickle.load(f)

with open('../models/feature_meta.pkl', 'rb') as f:
    meta = pickle.load(f)

print("✅ Pipeline loaded successfully")
print(f"Pipeline steps: {[step[0] for step in pipeline.steps]}")
print(f"Model: {meta['model_name']}")
print(f"Test Accuracy: {meta['test_accuracy']:.4f}")
print(f"Test F1 Score: {meta['test_f1']:.4f}")
```

---
## 📌 Step 2 — Test Pipeline with Sample Prediction

```python
# Sample input — Business traveler, loyal customer, business class
sample_input = {
    'gender': 'Male',
    'customer_type': 'Loyal Customer',
    'age': 40,
    'type_of_travel': 'Business travel',
    'class': 'Business',
    'flight_distance': 1500,
    'inflight_wifi_service': 4,
    'departure_and_arrival_time_convenience': 4,
    'ease_of_online_booking': 4,
    'gate_location': 3,
    'food_and_drink': 4,
    'online_boarding': 5,
    'seat_comfort': 5,
    'in_flight_entertainment': 4,
    'on_board_service': 4,
    'leg_room_service': 4,
    'baggage_handling': 4,
    'check_in_service': 4,
    'in_flight_service': 4,
    'cleanliness': 5,
    'departure_delay': 0,
    'arrival_delay': 0,
    'in_flight_wifi_service': 4,
}

# Feature engineering (same as Sprint 3)
service_cols = [
    'departure_and_arrival_time_convenience', 'ease_of_online_booking',
    'check_in_service', 'online_boarding', 'gate_location',
    'on_board_service', 'seat_comfort', 'leg_room_service',
    'cleanliness', 'food_and_drink', 'in_flight_service',
    'inflight_wifi_service', 'in_flight_entertainment', 'baggage_handling'
]

sample_input['total_delay']        = sample_input['departure_delay'] + sample_input['arrival_delay']
sample_input['delay_ratio']        = sample_input['arrival_delay'] / (sample_input['departure_delay'] + 1)
sample_input['avg_service_rating'] = np.mean([sample_input.get(c, 3) for c in service_cols])
sample_input['inflight_experience']= np.mean([sample_input['seat_comfort'],
                                               sample_input['in_flight_entertainment'],
                                               sample_input['in_flight_service'],
                                               sample_input['inflight_wifi_service']])
sample_input['ground_experience']  = np.mean([sample_input['check_in_service'],
                                               sample_input['gate_location'],
                                               sample_input['baggage_handling'],
                                               sample_input['ease_of_online_booking']])
sample_input['is_long_haul']       = int(sample_input['flight_distance'] > 1000)

sample_df = pd.DataFrame([sample_input])

# Predict
prediction = pipeline.predict(sample_df[meta['cat_cols'] + meta['num_cols']])[0]
proba      = pipeline.predict_proba(sample_df[meta['cat_cols'] + meta['num_cols']])[0]
result     = le_y.inverse_transform([prediction])[0]

print("=" * 50)
print("SAMPLE PREDICTION")
print("=" * 50)
print(f"Input: Business traveler, loyal, 40yo, Business class")
print(f"Prediction: {result.upper()}")
print(f"Confidence: {proba[prediction]*100:.1f}%")
print(f"Probabilities: {dict(zip(le_y.classes_, proba.round(3)))}")
```

## 📌 Step 4 — MLOps Practices

```python
# Logging predictions
import json
from datetime import datetime

log_entry = {
    'timestamp'  : datetime.now().isoformat(),
    'model'      : meta['model_name'],
    'prediction' : result,
    'confidence' : float(proba[prediction]),
    'input_age'  : sample_input['age'],
    'input_class': sample_input['class'],
}

print("📋 Sample Log Entry:")
print(json.dumps(log_entry, indent=2))

# Model metadata tracking
print("\n📊 Model Registry Entry:")
print(json.dumps({
    'model_name'   : meta['model_name'],
    'version'      : '1.0.0',
    'trained_on'   : datetime.now().strftime('%Y-%m-%d'),
    'test_accuracy': meta['test_accuracy'],
    'test_f1'      : meta['test_f1'],
    'pickle_path'  : 'models/final_model_pipeline.pkl',
    'framework'    : 'scikit-learn'
}, indent=2))
```
